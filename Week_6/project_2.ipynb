```python
import tkinter as tk
from tkinter import messagebox


class DeliverySystem:
    def __init__(self):
        # The prices for different locations and weights
        self.prices = {"Pau": {"heavy": 2000, "light": 1500},
                       "Epe": {"heavy": 5000, "light": 4000}}


    def calculate_cost(self, location, weight):
        location = location.capitalize()  


        if location not in self.prices:
            return None


        if weight >= 10:
            return self.prices[location]["heavy"]
        else:
            return self.prices[location]["light"]


class DeliveryGUI:
    def __init__(self, root):
        self.system = DeliverySystem()
        self.root = root
        self.root.title("Delivery Price Checker")
        self.root.geometry("420x260")


        tk.Label(root, text="Enter Delivery Location (PAU or Epe):", font=("Arial", 12)).pack(pady=5)
        self.location_entry = tk.Entry(root, width=30, font=("Arial", 12))
        self.location_entry.pack(pady=5)


        tk.Label(root, text="Enter Package Weight (in kg):", font=("Arial", 12)).pack(pady=5)
       self.weight_entry = tk.Entry(root, width=30, font=("Arial", 12))
        self.weight_entry.pack(pady=5)


        tk.Button(
            root,
            text="Check Price",
            command=self.calculate_price,
            font=("Arial", 12),
            bg="lightblue"
        ).pack(pady=15)


    def calculate_price(self):
        location = self.location_entry.get().strip()
        weight_text = self.weight_entry.get().strip()


        if not weight_text.isdigit():
            messagebox.showerror("Error", "Please enter a valid number for weight.")
            return


        weight = float(weight_text)


        cost = self.system.calculate_cost(location, weight)


        if cost is None:
            messagebox.showerror("Invalid Location", "Location must be either 'PAU' or 'Epe'.")
        else:
            messagebox.showinfo("Delivery Cost",
                f"Location: {location.capitalize()}\n"
                f"Weight: {weight} kg\n\n"
                f"You are to pay: ₦{cost}")
if __name__ == "__main__":
    root = tk.Tk()
    gui = DeliveryGUI(root)
    root.mainloop()
```
