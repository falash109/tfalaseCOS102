```python
import tkinter as tk
from tkinter import messagebox
import random


class EmployeeSystem:
    def __init__(self):
        self.employees = ["Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso",
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima" ]


        self.tasks = ["Loading", "Transporting", "Reveiwing Orders",
            "Customer Service", "Delivering Items"]


        self.attendance = {}


    def verify_employee(self, name):
        return name in self.employees


    def take_attendance(self, name):
        self.attendance[name] = "Present"


    def assign_task(self):
        return random.choice(self.tasks)


class EmployeeGUI:
    def __init__(self, root):
        self.system = EmployeeSystem() 
        self.root = root
        self.root.title("Employee Verification System")
        self.root.geometry("450x250")


        self.label = tk.Label(root, text="Enter Employee Name:", font=("Arial", 12))
        self.label.pack(pady=10)


        self.name_entry = tk.Entry(root, width=35, font=("Arial", 12))
        self.name_entry.pack(pady=5)


        self.btn = tk.Button(
            root,
            text="Verify & Assign Task",
            command=self.process_user,
            font=("Arial", 12),
            bg="lightblue"
        )
        self.btn.pack(pady=15)


    def process_user(self):
        name = self.name_entry.get().strip().title()


        if self.system.verify_employee(name):
            self.system.take_attendance(name) 
            task = self.system.assign_task() 


            messagebox.showinfo(
                "Access Granted",
                f"Welcome, {name}!\n\nAttendance Taken ✔\nToday's Task: {task}")
        else:
           messagebox.showwarning(
                "Access Denied",
                f"Sorry, {name} is not registered as an employee.")


if __name__ == "__main__":
    root = tk.Tk()
    gui = EmployeeGUI(root)
    root.mainloop() 
    
```
