```python
print("Equation Solver")
print("1. Quadratic Equation (ax² + bx + c = 0)")
print("2. Cubic Equation (ax³ + bx² + cx + d = 0)")
print("3. Quartic Equation (ax⁴ + bx³ + cx² + dx + e = 0)")

choice = int(input("Choose an option (1-3): "))

if choice == 1:
    print("Solving Quadratic Equation: ax² + bx + c = 0")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    
    discriminant = b**2 - 4*a*c
    
    if discriminant > 0:
        root1 = (-b + discriminant**0.5) / (2*a)
        root2 = (-b - discriminant**0.5) / (2*a)
        print("Roots are real and different")
        print("Root 1:", root1)
        print("Root 2:", root2)
    elif discriminant == 0:
        root = -b / (2*a)
        print("Roots are real and same")
        print("Root:", root)
    else:
        real = -b / (2*a)
        imaginary = (abs(discriminant)**0.5) / (2*a)
        print("Roots are complex")
        print("Root 1:", real, "+", imaginary, "i")
        print("Root 2:", real, "-", imaginary, "i")

elif choice == 2:
    print("Cubic Equation: ax³ + bx² + cx + d = 0")
    print("This is a cubic equation solver")
    print("For now, let's find one root by trial")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))
    
    print("Trying x = 0 to 5...")
    found = False
    for x in range(-10, 11):
        result = a*x**3 + b*x**2 + c*x + d
        if abs(result) < 0.001:
            print("Found a root:", x)
            found = True
            break
    
    if not found:
        print("Could not find integer root in range -10 to 10")

elif choice == 3:
    print("Quartic Equation: ax⁴ + bx³ + cx² + dx + e = 0")
    print("This is a quartic equation solver")
    print("For now, let's find one root by trial")
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))
    e = float(input("Enter e: "))
    
    print("Trying x = -5 to 5...")
    found = False
    for x in range(-5, 6):
        result = a*x**4 + b*x**3 + c*x**2 + d*x + e
        if abs(result) < 0.001:
            print("Found a root:", x)
            found = True
            break
    
    if not found:
        print("Could not find integer root in range -5 to 5")

else:
    print("Invalid choice")
```
