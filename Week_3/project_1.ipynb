```python
import tkinter as tk
from tkinter import messagebox

emp = {"John":"Logistics","Amina":"HR","Chidi":"Delivery","Fatima":"Finance"}

def check():
    n = name.get()
    d = dept.get()
    if n in emp:
        if emp[n] == d:
            messagebox.showinfo("Welcome", f"Hello {n}")
        else:
            messagebox.showerror("Error", f"Wrong dept! You are in {emp[n]}")
    else:
        messagebox.showerror("Error", "Not an employee")

root = tk.Tk()
root.title("GIG Verify")
tk.Label(root, text="Name").pack()
name = tk.Entry(root)
name.pack()
tk.Label(root, text="Dept").pack()
dept = tk.Entry(root)
dept.pack()
tk.Button(root, text="Verify", command=check).pack()
root.mainloop()
```
